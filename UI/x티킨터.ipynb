```python
text_num = 0
class GUI :
    def __init__(self):
        self.colcode = '#E3EFBB'
        if 'input_area' in globals():
            self.check_page()
        else:
            self.StartPage()
        self.win.mainloop() 
        
            
            
    def check_page(self) :
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)

        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = "사용자 이름을 입력해주세요.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        user_id = StringVar()
        tk.Entry(self.win, textvariable = user_id).pack()
                
        tk.Button(self.win, text="조회",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda : self.check_user(user_id)).pack(pady=15)
        tk.Button(self.win, text="이동", state = tk.DISABLED, width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.valid_page(user_id)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack()).pack(pady=15)

    def check_page_new(self, user_id) :
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)

        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = "사용자 이름을 입력해주세요.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        user_id_new = StringVar()
        a = tk.Entry(self.win,  textvariable = user_id_new)
        a.insert(0, user_id)
        a.pack()
        tk.Button(self.win, text="조회",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda : self.check_user(user_id_new)).pack(pady=15)
        tk.Button(self.win, text="이동", width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.valid_page(user_id_new)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack()).pack(pady=15)
        
    def valid_page(self, user_id) :
        user_num = self.check_user_ex(user_id)
        
        if user_num == -1 :
            self.check_page()
        else :    
            self.win.destroy()
        
            self.win = Tk()
            self.win.title("음성신호를 이용한 신원조회")
            self.win.geometry('400x450')
            self.win.resizable(False, False)
            self.win.configure(bg=self.colcode)        
            tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
            tk.Label(self.win, text=user_id.get() + "님 어서오세요", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
            
            tk.Button(self.win, text="인증하기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.record_page(user_id)).pack(pady=15)
            new_id = user_id
            tk.Button(self.win, text="음성 재등록",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.register01_page1(new_id)).pack(pady=15)        
            tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack()).pack(pady=15)        

    def register00_page(self) :
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)
        
        
        tk.Label(self.win, text = "신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = "등록할 사용자 이름을 입력해주세요.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        
        
        new_id = StringVar() # 신규이용자 이름         
        
        tk.Entry(self.win, textvariable = new_id).pack()
        
        tk.Button(self.win, text="사용자 신규 등록",width="12",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.new_user(new_id)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="12",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack()).pack(pady=15)
        
    def register01_page(self, new_id):
        self.win.destroy()
        dir_Data_path = "./User_Data/" + new_id
        dir_Mfcc_path = "./User_Mfcc/" + new_id
        if os.path.exists(dir_Data_path) :
            shutil.rmtree(dir_Data_path)
        if os.path.exists(dir_Mfcc_path) :
            shutil.rmtree(dir_Mfcc_path)    
            
        os.mkdir(dir_Data_path)
        os.mkdir(dir_Mfcc_path)    
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)

        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = new_id + "님 반갑습니다.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        
        tk.Button(self.win, text="음성등록",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.record_user(new_id)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack_Delete(new_id)).pack(pady=15)
        
    def register01_page1(self, new_id):
        self.win.destroy()
        new_id = new_id.get()
        dir_Data_path = "./User_Data/" + new_id
        dir_Mfcc_path = "./User_Mfcc/" + new_id
        if os.path.exists(dir_Data_path) :
            shutil.rmtree(dir_Data_path)
        if os.path.exists(dir_Mfcc_path) :
            shutil.rmtree(dir_Mfcc_path)    
            
        os.mkdir(dir_Data_path)
        os.mkdir(dir_Mfcc_path)    
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)

        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = new_id + "님 반갑습니다.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        
        tk.Button(self.win, text="음성등록",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.record_user(new_id)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack_Delete(new_id)).pack(pady=15)
        
    def register02_page(self, new_id) :
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)
        
        
        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = "등록할 사용자 문장을 입력해주세요.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        
        new_sentence = StringVar() #신규이용자가 사용할 문장
        
        tk.Entry(self.win, textvariable = new_sentence).pack()
                  
        tk.Button(self.win, text="사용자 문장 등록",width="12",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.user_sentence(new_id, new_sentence)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="12",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack_Delete(new_id)).pack(pady=15)
                
    def user_sentence(self, new_id, new_sentence) :  
        
        new_sentence = new_sentence.get()
        
        f = open(new_id + '_PrimarySentnce.csv','w', newline = "") 
        wr = csv.writer(f)
        wr.writerow([new_sentence])
        f.close()
        
        self.Record_user_Re1(new_id, new_sentence)

    def record_user(self, new_id):
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)

        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = new_id + "님 반갑습니다.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        
        self.text = tk.StringVar()
        
        
        tk.Button(self.win, text="녹음",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Voice_Record_Register(new_id)).pack(pady=15)
        text_label = tk.Label(self.win, textvariable = self.text)
        tk.Button(self.win, text="다음문장",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.server_text(new_id)).pack(pady=15)
        text_label.pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack_Delete(new_id)).pack(pady=15)
        # 사용자의 목소리를 녹음하는 기능 
        
    def Record_user_Re1(self, new_id, new_sentence) : 
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)
        
        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = new_id + "님 반갑습니다.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        
        self.text = tk.StringVar()
        
        tk.Button(self.win, text="녹음",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Voice_Record_Register(new_id)).pack(pady=15)
        tk.Button(self.win, text="이동",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Record_user_Re2(new_id, new_sentence)).pack(pady=15)
        text_label = tk.Label(self.win, text = "##" + new_sentence + "##").pack(pady=15)
        tk.Button(self.win, text="완료",state = tk.DISABLED, width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Success(new_id)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack_Delete(new_id)).pack(pady=15)
    def Record_user_Re2(self, new_id, new_sentence) : 
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)
        
        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = new_id + "님 반갑습니다.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        
        self.text = tk.StringVar()
        
        tk.Button(self.win, text="녹음",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Voice_Record_Register(new_id)).pack(pady=15)
        tk.Button(self.win, text="이동",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Record_user_Re3(new_id, new_sentence)).pack(pady=15)
        text_label = tk.Label(self.win, text = "##" + new_sentence + "##").pack(pady=15)
        tk.Button(self.win, text="완료",state = tk.DISABLED, width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Success(new_id)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack_Delete(new_id)).pack(pady=15)
    def Record_user_Re3(self, new_id, new_sentence) : 
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)
        
        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = new_id + "님 반갑습니다.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        
        self.text = tk.StringVar()
        
        tk.Button(self.win, text="녹음",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Voice_Record_Register(new_id)).pack(pady=15)
        tk.Button(self.win, text="이동",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Record_user_Re4(new_id, new_sentence)).pack(pady=15)
        text_label = tk.Label(self.win, text = "##" + new_sentence + "##").pack(pady=15)
        tk.Button(self.win, text="완료",state = tk.DISABLED, width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Success(new_id)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack_Delete(new_id)).pack(pady=15)
    def Record_user_Re4(self, new_id, new_sentence) : 
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)
        
        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = new_id + "님 반갑습니다.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        
        self.text = tk.StringVar()
        
        tk.Button(self.win, text="녹음",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Voice_Record_Register(new_id)).pack(pady=15)
        tk.Button(self.win, text="이동",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Record_user_Re5(new_id, new_sentence)).pack(pady=5)
        text_label = tk.Label(self.win, text = "##" + new_sentence + "##").pack(pady=15)
        tk.Button(self.win, text="완료",state = tk.DISABLED, width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Success(new_id)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack_Delete(new_id)).pack(pady=15)

    def Record_user_Re5(self, new_id, new_sentence) : 
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)
        
        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = new_id + "님 반갑습니다.", font=('Helvetica', 16, "bold")).pack(side="top", fill="x", pady=5)
        
        self.text = tk.StringVar()
        
        tk.Button(self.win, text="녹음",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Voice_Record_Register(new_id)).pack(pady=15)
        text_label = tk.Label(self.win, text = "##" + new_sentence + "##").pack(pady=15)
        tk.Button(self.win, text="완료",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Success(new_id)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack_Delete(new_id)).pack(pady=15)  

                
    def server_text(self,new_id) :   # 서버에서 지정한 문장 출력 
        d = open('sentence.csv','r', encoding='cp949')
        rdr = csv.reader(d)
        sentence = []
        for line in rdr:
            sentence.append(line)
        d.close()
        global text_num         
        
        if text_num == 1 :
            self.register02_page(new_id)

        elif text_num == 0 :
            text = "## 안녕하세요 저는 " + new_id + "입니다 ##"
            self.text.set(text)
            
        else :
            self.text.set(sentence[text_num])

        print(text_num)
        text_num += 1
            
    def new_user(self, new_id) :
        
        f = open('./user_info.csv','r', encoding='cp949')
        rdr = csv.reader(f)
        user_list = []
        line_num = 0
        user_list1 = []        
        id=[]        
        for line in rdr:

            user_list.append(line)
        f.close()        
        
        num = IntVar()
        new_id = new_id.get()
        n = 0
        while n < len(user_list) :
            if new_id == user_list[n][0] :
                print("이미 등록된 사용자입니다.")
                messagebox.showinfo("결과창" ,"이미 등록된 사용자입니다.")
                self.register00_page()
                return -1
            
            elif n == len(user_list)-1 :
                user_list.append(new_id)
                self.register01_page(new_id)
                messagebox.showinfo("결과창" ,new_id + "님 반갑습니다.")
                
                return user_list
            n += 1
        
    def record_page(self, user_id) : 
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)         
        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = user_id.get() + "님 어서오세요", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        
        user_num = self.check_user_ex(user_id)                
        
        tk.Button(self.win, text="인증",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Voice_Record(user_id, user_num)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack()).pack(pady=15)

    def record_page_Re(self, user_id) : 
        self.win.destroy()
        
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)         
        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Label(self.win, text = user_id + "님 어서오세요", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        
        user_num = self.check_user_ex(user_id)                
        
        tk.Button(self.win, text="인증",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.Voice_Record(user_id, user_num)).pack(pady=15)
        tk.Button(self.win, text="나가기",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.goBack()).pack(pady=15)  
        
    def check_user(self, user_id) :
        user_id = user_id.get()
        user_id1 = user_id + "\n"
        f = open('user_info.csv', 'r',encoding='cp949')
        line_num = 0
        line = f.readline()
        id=[]
        while line:
            id += [[line_num, line]]
            line = f.readline()
            line_num += 1
        f.close()
        
        n = 0     
        while n < len(id) :
            if user_id1 == id[n][1] :   
                messagebox.showinfo("결과창" ,user_id + "님 확인되었습니다.")
                self.check_page_new(user_id)      
                return n
            elif n == len(id)-1 :
                messagebox.showinfo("결과창" ,"다시 입력하여 주십시오.")
                return -1
            n += 1 
        
        messagebox.showinfo("결과창" ,user_id + "님 확인되었습니다.")
        self.check_page_new()        

    def check_user_ex(self, user_id) :
        user_id = user_id.get()
        user_id1 = user_id + "\n"
        
        f = open('user_info.csv', 'r',encoding='cp949')
        line_num = 0
        line = f.readline()
        id=[]
        while line:
            id+=[[line_num, line]]
            line = f.readline()
            line_num += 1
        f.close()
        
        n = 0
        
        while n < len(id) :
            if user_id1 == id[n][1] :     
                return n
            elif n == len(id)-1 :
                return -1
            n += 1 
    
    def switch_frame(self, frame_class) :
        new_frame = frame_class(self)
        if self._frame is not None:
            self._frame.destroy()
        self._frame = new_frame
        self._frame.pack()
        
    def goBack(self) :
        self.win.destroy()
        self.StartPage()
        
    def goBack_Delete(self, new_id) :
        shutil.rmtree("./User_Data/"+str(new_id + "/"))
        shutil.rmtree("./User_Mfcc/"+str(new_id + "/"))
        
        self.win.destroy()
        self.StartPage()
    
    def Success(self, new_id) :
        
        f = open('user_info.csv','a', newline = "") 
        wr = csv.writer(f)
        wr.writerow([new_id])
        f.close()

        messagebox.showinfo("결과창" ,"등록이 완료 되었습니다.")
        self.win.destroy()
        self.StartPage()
        
    def StartPage(self) :
        self.win = Tk()
        self.win.title("음성신호를 이용한 신원조회")
        self.win.geometry('400x450')
        self.win.resizable(False, False)
        self.win.configure(bg=self.colcode)
        global text_num
        text_num = 0
        tk.Label(self.win, text="신원조회 프로그램", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tk.Button(self.win, text="조회",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.check_page()).pack(pady=15)
        tk.Button(self.win, text="등록",width="8",height="2",font=('Helvetica', 12, "bold"),
                  command=lambda: self.register00_page()).pack(pady=15)
        
    def Voice_Record(self, user_id, user_num) :
        
        user_id = user_id.get()
        
        FORMAT = pyaudio.paInt16
        CHANNELS = 1
        RATE = 44100 #비트레이트 설정
        CHUNK = 1024
        RECORD_SECONDS = 5 # 녹음할 시간 설정 5초 너무 짧은 느낌이 들기도 하네요
        WAVE_OUTPUT_FILENAME = "./User_Data/" + user_id + "/"
        FILE_NAME = user_id + "_"

        files = os.listdir(WAVE_OUTPUT_FILENAME)
        wave_count = 1
    
        #폴더 이름과 그 폴더에 속하는 파일 갯수 출력
        for wav in files: 
            if not wav.endswith(".wav"):continue
            else: wave_count += 1


        WAVE_OUTPUT_FILENAME = WAVE_OUTPUT_FILENAME + FILE_NAME + str(wave_count)+".wav"
    
        p = pyaudio.PyAudio() # 오디오 객체 생성

        stream = p.open(format=FORMAT, # 16비트 포맷
                    channels = CHANNELS,
                    rate=RATE, #비트레이트
                    input=True,
                    frames_per_buffer=CHUNK) # CHUNK만큼 버퍼가 쌓인다.
        messagebox.showinfo("결과창" ,user_id + "님 5초간 녹음을 시작하겠습니다.")
        print("Start to Record the Audio.")

        frames = [] # 음성 데이터를 채우는 공간

        for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)): 
            data = stream.read(CHUNK)
            frames.append(data)

        print("Recording is Finished.")
        messagebox.showinfo("결과창" ,user_id + "님 녹음이 끝났습니다.")
        stream.stop_stream() # 스트림닫기
        stream.close() # 스트림 종료
        p.terminate() # 오디오객체 종료
    
        wf = wave.open( WAVE_OUTPUT_FILENAME, 'wb') 
        # WAVE_OUTPUT_FILENAME의 파일을 열고 데이터를 쓴다.
        wf.setnchannels(CHANNELS)
        wf.setsampwidth(p.get_sample_size(FORMAT))
        wf.setframerate(RATE)
        wf.writeframes(b''.join(frames))
        wf.close()
        
        user_feature_list = os.listdir("./User_Data/") 
        
        save_path = "./User_Data/"
        user_list = os.listdir(save_path) # 경로에 있는 폴더 리스트
        user_list.sort() # 정렬
        r = speech_r.Recognizer()  
            
        hop_length = 256
        n_mels = 40
        n_fft = 512
        sr = 16000

            
        y, sr = librosa.load(WAVE_OUTPUT_FILENAME , sr=16000, mono=True, duration=10)
        y = y[0:int(5.0 * sr)]
                
        stft = librosa.stft(y=y, n_fft=n_fft, hop_length=hop_length, win_length=None, window='hann',
        center=False, dtype=np.complex64)
        
        abs_STFT = np.abs(stft)
    
        mel = librosa.filters.mel(sr=sr, n_fft=n_fft, n_mels=n_mels)
        mel /= np.max(mel, axis=-1)[:, None]
        mel_S = np.dot(mel, abs_STFT)

        mel_spec_test = 20 * np.log10(np.maximum(mel_S, 1e-5))
    
        num_ceps = 19
        cep_lifter = 22
    
        mfcc = dct(mel_spec_test.T, type=2 ,axis=1, norm='ortho')[:, 1 : (num_ceps + 1)]
        (nframes, ncoeff) = mfcc.shape

        n = np.arange(ncoeff)
    
    
        # to de-emphasize higher MFCCs (for Noise)
        lift = 1 + (cep_lifter / 2) * np.sin(np.pi * n / cep_lifter)
        mfcc_lifted = mfcc * lift
            
                #mfcc_normalized = mfcc_lifted -(np.mean(mfcc_lifted, axis=0) + 1e-8)
    
        feature_save_path = "./User_Mfcc/"+str(user_list[user_num]+"/")


        feature_save_path = "./User_Mfcc/" + str(user_list[user_num]+"/")
        feature_files = os.listdir(feature_save_path)
        feature_files = [file for file in feature_files if file.endswith(".npy")]
        feature_result_name = str(user_list[user_num]) +"_"+str(len(feature_files))

        final_name = str(feature_save_path) + str(feature_result_name)
        print(feature_result_name + "' File Write Finish !")
        print(" Saved File Name : "+ feature_result_name)
        np.save(final_name, mfcc_lifted) # x_save.npy
            
        model = load_model("./모델/model_1.h5")

         
        XXX_test = np.load(final_name + ".npy")
        XXX_test = np.expand_dims(XXX_test, -1)

        label = [0 for i in range(11)]
        label[user_num] = 1

        YY_test = []
        for i in range(len(XXX_test)):
            YY_test.append(label)


        XXX_test = np.array(XXX_test)
        YY_test = np.array(YY_test)


        A = model.evaluate(XXX_test, YY_test)
    
        if A[1] > 0.9 :
            messagebox.showinfo("결과창" ,user_id + "님 어서오세요!") #메시지 박스를 띄운다.
        elif A[1] > 0.5 : 
            messagebox.showinfo("결과창" ,user_id + "님? 어서오세요?") #메시지 박스를 띄운다.
        else :
            self.goBack()
            
    def Voice_Record_Register(self, new_id) :
        
        user_id = new_id   
        
        FORMAT = pyaudio.paInt16
        CHANNELS = 1
        RATE = 44100 #비트레이트 설정
        CHUNK = 1024
        RECORD_SECONDS = 5 # 녹음할 시간 설정 5초 너무 짧은 느낌이 들기도 하네요
        WAVE_OUTPUT_FILENAME = "./User_Data/" + user_id + "/"
        FILE_NAME = user_id + "_"

        files = os.listdir(WAVE_OUTPUT_FILENAME)
        wave_count = 1
    
        #폴더 이름과 그 폴더에 속하는 파일 갯수 출력
        for wav in files: 
            if not wav.endswith(".wav"):continue
            else: wave_count += 1


        WAVE_OUTPUT_FILENAME = WAVE_OUTPUT_FILENAME + FILE_NAME + str(wave_count)+".wav"
    
        p = pyaudio.PyAudio() # 오디오 객체 생성

        stream = p.open(format=FORMAT, # 16비트 포맷
                    channels = CHANNELS,
                    rate=RATE, #비트레이트
                    input=True,
                    frames_per_buffer=CHUNK) # CHUNK만큼 버퍼가 쌓인다.
        messagebox.showinfo("결과창" ,user_id + "님 5초간 녹음을 시작하겠습니다.")
        print("Start to Record the Audio.")

        frames = [] # 음성 데이터를 채우는 공간

        for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)): 
            data = stream.read(CHUNK)
            frames.append(data)

        print("Recording is Finished.")
        messagebox.showinfo("결과창" ,user_id + "님 녹음이 끝났습니다.")
        stream.stop_stream() # 스트림닫기
        stream.close() # 스트림 종료
        p.terminate() # 오디오객체 종료
    
        wf = wave.open( WAVE_OUTPUT_FILENAME, 'wb') 
        # WAVE_OUTPUT_FILENAME의 파일을 열고 데이터를 쓴다.
        wf.setnchannels(CHANNELS)
        wf.setsampwidth(p.get_sample_size(FORMAT))
        wf.setframerate(RATE)
        wf.writeframes(b''.join(frames))
        wf.close()
        
        user_feature_list = os.listdir("./User_Data/") 
        
        save_path = "./User_Data/"
        user_list = os.listdir(save_path) # 경로에 있는 폴더 리스트
        user_list.sort() # 정렬
        r = speech_r.Recognizer()  
            
        hop_length = 256
        n_mels = 40
        n_fft = 512
        sr = 16000

            
        y, sr = librosa.load(WAVE_OUTPUT_FILENAME , sr=16000, mono=True, duration=10)
        y = y[0:int(5.0 * sr)]
                
        stft = librosa.stft(y=y, n_fft=n_fft, hop_length=hop_length, win_length=None, window='hann',
        center=False, dtype=np.complex64)
        
        abs_STFT = np.abs(stft)
    
        mel = librosa.filters.mel(sr=sr, n_fft=n_fft, n_mels=n_mels)
        mel /= np.max(mel, axis=-1)[:, None]
        mel_S = np.dot(mel, abs_STFT)

        mel_spec_test = 20 * np.log10(np.maximum(mel_S, 1e-5))
    
        num_ceps = 19
        cep_lifter = 22
    
        mfcc = dct(mel_spec_test.T, type=2 ,axis=1, norm='ortho')[:, 1 : (num_ceps + 1)]
        (nframes, ncoeff) = mfcc.shape

        n = np.arange(ncoeff)
    
    
        # to de-emphasize higher MFCCs (for Noise)
        lift = 1 + (cep_lifter / 2) * np.sin(np.pi * n / cep_lifter)
        mfcc_lifted = mfcc * lift
            
                #mfcc_normalized = mfcc_lifted -(np.mean(mfcc_lifted, axis=0) + 1e-8)
    
        feature_save_path = "./User_Mfcc/"+str(new_id)+"/"


        feature_save_path = "./User_Mfcc/" + str(new_id)+"/"
        feature_files = os.listdir(feature_save_path)
        feature_files = [file for file in feature_files if file.endswith(".npy")]
        feature_result_name = str(new_id) +"_"+str(len(feature_files))

        final_name = str(feature_save_path) + str(feature_result_name)
        print(feature_result_name + "' File Write Finish !")
        print(" Saved File Name : "+ feature_result_name)
        np.save(final_name, mfcc_lifted) # x_save.npy            

```
